```python
import pandas as pd

test_data = pd.read_csv("./csv/test.csv")
train_data = pd.read_csv("./csv/train.csv")
test_data
```

```python
train_data
```

```python
# Model  X  ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd'] ==> Y (SalePrice)
y = train_data["SalePrice"]
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = train_data[features]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(X,y,train_size=0.8,test_size=0.2, random_state=1)

```

```python
from sklearn.ensemble import RandomForestRegressor

m1 = RandomForestRegressor(n_estimators=50, random_state=0)
m2 = RandomForestRegressor(n_estimators=100, random_state=0)
m3 = RandomForestRegressor(n_estimators=100, random_state=0, criterion="absolute_error")
m4 = RandomForestRegressor(n_estimators=200, random_state=0, min_samples_split=20)
m5 = RandomForestRegressor(n_estimators=100, random_state=0, max_depth=7)

models = [m1,m2,m3,m4,m5]
```

```python
m1.fit(X_train,y_train)
```

```python
preds = m2.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_valid, preds)
```

```python
for i in models:
    i.fit(X_train,y_train)
    preds = i.predict(X_valid)
    outs = mean_absolute_error(y_valid, preds)
    print(outs)
```
