```python
# output target
y = home_data.Price
# input fetaures array
input_feat = ["Rooms", "Bathroom", "Car", "YearBuilt", "Landsize", "Lattitude", "Longtitude","BuildingArea","Propertycount"]
# input feature
X = home_data[input_feat]

def result(nodes):
    # create a model
    new_model  = DecisionTreeRegressor(max_leaf_nodes=nodes, random_state=1)
    # Split the training and test data
    X_train, X_test, y_train, y_test  = train_test_split(X,y, random_state=1)

    # fit or train the model
    new_model.fit(X_train,y_train)
    # predict the model with test data
    pred = new_model.predict(X_test)
    return mean_absolute_error(y_test,pred)

for node in [5,50,200,300,500,1000]:
    print(result(node))


```
